### RavenStack SaaS Dataset — Initial Data Exploration

This notebook explores the structure, relationships, and key fields of a multi-table SaaS dataset to understand available metrics and identify analytical opportunities.

```python
import pandas as pd

accounts = pd.read_csv("../data/accounts.csv")
subscriptions = pd.read_csv("../data/subscriptions.csv")
feature_usage = pd.read_csv("../data/feature_usage.csv")
support_tickets = pd.read_csv("../data/support_tickets.csv")
churn_events = pd.read_csv("../data/churn_events.csv")

dataframes = [
    ("accounts", accounts),
    ("subscriptions", subscriptions),
    ("feature_usage", feature_usage),
    ("support_tickets", support_tickets),
    ("churn_events", churn_events)
]

for name, df in dataframes:
    print(f"DataFrame: {name}\n")
    print(df.head(), "\n")
    df.info()
    print(df.describe(include="all"), "\n\n")
```

```python
accounts.account_id.nunique()
subscriptions.account_id.nunique()
feature_usage.subscription_id.nunique()
```

Initial Observations

- Revenue data is captured at the subscription level (MRR/ARR)

- Usage data is highly granular and time-based

- Churn is modeled as discrete events with behavioral context

- Support data enables operational workload analysis
